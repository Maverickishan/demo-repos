## Advanced Options

Beyond local editing two things are quite important: editing the notebooks in the markdown format and running Jupyter remotely.
The latter matters when we want to run the code on a faster server.
The former matters since Jupyter's native ipynb format stores a lot of auxiliary data that is
irrelevant to the content,
mostly related to how and where the code is run.
This is confusing for Git, making
reviewing contributions very difficult.
Fortunately there is an alternative---native editing in the markdown format.

### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

### Running Jupyter Notebooks on a Remote Server

Sometimes, you may want to run Jupyter notebooks on a remote server and access it through a browser on your local computer. If Linux or macOS is installed on your local machine (Windows can also support this function through third-party software such as PuTTY), you can use port forwarding:

```
ssh myserver -L 8888:localhost:8888
```

The above string `myserver` is the address of the remote server.
Then we can use http://localhost:8888 to access the remote server `myserver` that runs Jupyter notebooks. We will detail on how to run Jupyter notebooks on AWS instances
later in this appendix.

### Timing

We can use the `ExecuteTime` plugin to time the execution of each code cell in Jupyter notebooks.
Use the following commands to install the plugin:

```
pip install jupyter_contrib_nbextensions
jupyter contrib nbextension install --user
jupyter nbextension enable execute_time/ExecuteTime
```

````python
import yfinance as yf

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt



##Input Data

# Step 1: Define stocks and fetch data from Yahoo Finance

stocks = ['RELIANCE.NS', 'TCS.NS', 'HDFCBANK.NS', 'INFY.NS', 'HINDUNILVR.NS']



# risk-free rate

rf = 0.04



# Fetch 5 years of data of daily closing price

data = yf.download(stocks, period="5y")['Close']
````

````python
returns = data.pct_change().dropna()
````

````python
mu = returns.mean()*252
sigma = returns.cov() *252
std_dev = np.sqrt(np.diag(sigma))
````

````python
plt.figure(figsize=(10,6))
plt.scatter(std_dev, mu, color= 'black', marker = 'o', s=100, label='Individual')
plt.xlabel('risk(Std Dev)')
plt.ylabel('Return')
plt.title('Return and Risk of Assets')
plt.legend()
````

````python
#opportunity set
#number of assets
num_assets = len(stocks)
#generate random portfolios of opportunity set (non-negative weights) using Numpy
num_portfolios = 100000
#Generate random  weights
weights = np.random.random((num_portfolios,num_assets))
weights = weights/ np.sum(weights, axis=1,keepdims= True)
#compute portfolio returns: W* mu (dot product)
port_returns = np.dot(weights, mu)

#compute portfolio standrad deviation : sqrt (w* sigma * w^t)
port_stds = np.sqrt(np.einsum('ij,jk,ik->i',weights,sigma,weights))
#Store results in array(return, std)
results = np.vstack((port_returns,port_stds))

#plot opportunity set
plt.figure(figsize=(10,6))
plt.scatter(results[1,:], results[0,:], c = results[0,], marker='o', cmap='YlOrRd',s=10, alpha=0.3, label='opportunity set')
plt.scatter(std_dev, mu, color= 'black', marker = 'o', s=100, label='Individual')
plt.xlabel('risk(Std Dev)')
plt.ylabel('Return')
plt.title('Return and Risk of Assets')
plt.legend()

````

````python
# ___ global minimum variance protfolio__
 #vector of one
ones= np.ones(num_assets)

#compute inverse convariance matrix
sigma_inv = np.linalg.inv(sigma)

#GMV weights
w_gmv = sigma_inv @ ones / (ones.T @ sigma_inv @ones)

# GMV expected retuen
mu_gmv = w_gmv @ mu
#GMV risk (std dev)
sigma_gmv = np.sqrt(w_gmv.T @ sigma @w_gmv)

print("GMV Portfolio Weights:")
for stock,weight in zip(stocks, w_gmv):
  print(f"{stock}: {weight:.4f}")

print(f"\nGMV Expected  Return: {mu_gmv:.4f}")
print(f"\nGMV Risk(std div): {sigma_gmv:.4f}")

# plot GMV on the opportunity set
plt.figure(figsize=(10,6))
plt.scatter(results[1,:], results[0,:], c = results[0,], marker='o', cmap='YlOrRd',s=10, alpha=0.3, label='opportunity set')
plt.scatter(std_dev, mu, color= 'black', marker = 'o', s=100, label='Individual')
plt.scatter(sigma_gmv, mu_gmv, color= 'green', marker = 'o', s=100, label='GMV_portfoloio')
plt.xlabel('risk(Std Dev)')
plt.ylabel('Return')
plt.title('Portfolio Opportunity set with GMU')
plt.legend()
plt.show()
````

````python
# ___ Tangency Portfolio__
rf = 0.04

# vector of ones
ones = np.ones(num_assets)

#Excess return (mu - rf*1)
excess_mu = mu - rf*ones

# tangency Weights
w_tan = sigma_inv @ excess_mu / (ones.T @ sigma_inv @ excess_mu)

# tangency porfolio return
mu_tan = w_tan @ mu

#tangency portfolio risk (std dev)
sigma_tan = np.sqrt(w_tan.T @ sigma @ w_tan)

print("\n Tangency protfolio weights:")
for stock, weight in zip(stocks, w_tan):
  print(f"{stock}: {weight:.4f}")

print(f"\nTangency Portfolio Expected Return: {mu_tan:.4f}")
print(f"Tangency Portfolio Risk (Std Dev ):{sigma_tan:.4f}")
print(f"Tangency Portfolio Sharpr Ratio: {(mu_tan-rf)/sigma_tan:.4f}")

#plot Opportunity Set + GMV + Tnagency portfolio
plt.figure(figsize=(10,6))
plt.scatter(results[1,:], results[0,:], c = results[0,], marker='o', cmap='YlOrRd',s=10, alpha=0.3, label='opportunity set')
plt.scatter(std_dev, mu, color= 'black', marker = 'o', s=100, label='Individual')
plt.scatter(sigma_gmv, mu_gmv, color= 'green', marker = 'o', s=100, label='GMV_portfoloio')
plt.scatter(sigma_tan, mu_tan, color='blue', marker = 'o', s=100, label='Tangency Portfolio')

# Capital Market line (CML)
x_val = np.linspace(0, max(results[1,:]), 100)
y_vals= rf + ((mu_tan-rf)/ sigma_tan)* x_val
plt.plot(x_val,y_vals, color='blue', linestyle='--', label='capital maekt line')

plt.xlabel('risk(Std Dev)')
plt.ylabel('Return')
plt.title('Portfolio Opportunity set with GMU')
plt.legend()
plt.show()
````

````python
from ast import Return
# -- optimal Portfolio for Desired Return

mu_star = 0.13

# compute A B C  D

A = ones.T @ sigma_inv @ mu
B= mu.T @ sigma_inv @ mu
C = ones.T @ sigma_inv @ ones
D = B*C - A**2

# lagranges multiplier

lam = (C * mu_star - A) / D
gam = (B - A*mu_star)/D

#optimal weights
w_opt = sigma_inv@(lam*mu + gam*ones)

#return & Risk
mu_opt = w_opt @ mu
sigma_opt = np.sqrt(w_opt.T @ sigma @ w_opt)

print(f"\nOptimal Portfolio (Target Return ={mu_star:.2f}):")
for stock, weight in zip(stocks, w_opt):
  print(f"{stock}: {weight:.4f}")

print(f"\nOptimal Portfolio Expected Return: {mu_opt:.4f}")
print(f"Optimal Portfolio Risk (Std Dev ):{sigma_opt:.4f}")

#--- Plot with GMV & Tangency____

plt.figure(figsize=(10,6))
plt.scatter(results[1,:], results[0,:], c = results[0,], marker='o', cmap='YlOrRd',s=10, alpha=0.3, label='opportunity set')
plt.scatter(std_dev, mu, color= 'black', marker = 'o', s=100, label='Individual')
plt.scatter(sigma_gmv, mu_gmv, color= 'green', marker = 'o', s=100, label='GMV_portfoloio')
plt.scatter(sigma_tan, mu_tan, color='blue', marker = 'o', s=100, label='Tangency Portfolio')
plt.scatter(sigma_opt, mu_opt, color='green', marker = '*', s=100, label='Optimal Portfolio({mu_star:.0%}Return)')
plt.xlabel('risk(Std Dev)')
plt.ylabel('Return')
plt.title('Portfolio Opportunity set with GMU')
plt.legend()
plt.show()


````
